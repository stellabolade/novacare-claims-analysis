CLEANED RECORD (REMOVING DUPLICATES)

```python


# Turn 'Denied' into 1 and 'Paid' into 0
df['is_denied'] = (df['claim_status'] == 'Denied').astype(int)
```

```python
df_unique = df[df['is_duplicate_claim'] == 0].copy()


df_unique['is_denied'] = (df_unique['claim_status'] == 'Denied').astype(int)


total_unique_claims = len(df_unique)
total_denied = df_unique['is_denied'].sum()
denial_rate = df_unique['is_denied'].mean()


print(f"Total Unique Claims: {total_unique_claims}")
print(f"Total Unique Denials: {total_denied}")
print(f"Overall Denial Rate: {denial_rate:.2%}")
```

```python
# Calculate the overall rate first
overall_rate = df['is_denied'].mean()
print(f"Overall Denial Rate: {overall_rate:.2%}")

```

DENIAL RATE BY DEPARTMENT

```python
dept_denial_rate = df_unique.groupby('department')['is_denied'].agg(
    total_claims='count',
    total_denied='sum',
    denial_rate='mean'
).sort_values(by='denial_rate', ascending=False)

print(dept_denial_rate)
```

DENIAL RATE BY PROVIDER

```python
# Grouping to see the stats by category
provider_denial_rate = df_unique.groupby('provider_name')['is_denied'].agg(
    total_claims='count',
    total_denied='sum',
    denial_rate='mean'
).sort_values(by='denial_rate', ascending=False)

print(provider_denial_rate)
```

DENIAL RATE BY PAYER TYPE

```python
payer_denial_rate = df_unique.groupby('payer_type')['is_denied'].agg(
    total_claims='count',
    total_denied='sum',
    denial_rate='mean'
).sort_values(by='denial_rate', ascending=False)

print(payer_denial_rate)
```

DENIAL RATE BY ICD10

```python
ICD10_denial_rate = df_unique.groupby('icd10_primary')['is_denied'].agg(
    total_claims='count',
    total_denied='sum',
    denial_rate='mean'
).sort_values(by='denial_rate', ascending=False)

print(ICD10_denial_rate)
```

REVENUE LEAKAGE: Money hospitals should have earned from patient care, but did not receive because claims were denied, delayed, or rejected....Duplicate denied claims were identified and removed prior to analysis to prevent overestimation of revenue leakage.”


```python
#Filter for claims that are Denied AND NOT duplicates
true_denied_claims = df[(df['claim_status'] == 'Denied') & (df['is_duplicate_claim'] == 0)]

#Calculate the ACCURATE Total Leakage
true_revenue_leakage = true_denied_claims['billed_amount'].sum()

#Calculate the "False" Leakage
false_total_leakage = df[df['claim_status'] == 'Denied']['billed_amount'].sum()
inflation_amount = false_total_leakage - true_revenue_leakage

print(f"True Revenue Leakage: ${true_revenue_leakage:,.2f}")
print(f"Amount Overstated by Duplicates: ${inflation_amount:,.2f}")
```

```python
Df_unique = df[df['is_duplicate_claim'] == 0].copy()
df_denied = df_unique[df_unique['claim_status'] == 'Denied']

# Group by department and calculate the metrics
leakage_by_department = (
    df_denied.groupby('department')
    .agg(
        total_leakage=('billed_amount', 'sum'),
        total_claims=('billed_amount', 'count')
    )
    .sort_values(by='total_leakage', ascending=False)
    .reset_index()
)

print(leakage_by_department)
```

LEAKED BY PROVIDER

```python
df_unique = df[df['is_duplicate_claim'] == 0].copy()
df_denied = df_unique[df_unique['claim_status'] == 'Denied']


leakage_by_provider = (
    df_denied.groupby('provider_name')
    .agg(
        total_leakage=('billed_amount', 'sum'),
        total_claims=('billed_amount', 'count')
    )
    .sort_values(by='total_leakage', ascending=False)
    .reset_index()
)
print(leakage_by_provider)
```

LEAKAGE BY PAYER TYPE

```python
df_denied = df[df['claim_status'] == 'Denied']


leakage_by_payer = (
    df_denied.groupby('payer_type')
    .agg(
        total_leakage=('billed_amount', 'sum'),
        total_claims=('billed_amount', 'count')
    )
    .sort_values(by='total_leakage', ascending=False)
    .reset_index()
)
print(leakage_by_payer)
```

CORRELATION ANALYSIS


```python

df['is_denied'] = (df['claim_status'] == 'Denied').astype(int)


print("Correlation Analysis")
corr_table = df[['billed_amount', 'claim_processing_days', 'is_denied']].corr()
display(corr_table)

print("\n")


means_table = df.groupby('is_denied')[['billed_amount', 'claim_processing_days']].mean()
display(means_table)
```

DENIAL REASON ANALYSIS

```python
#Calculate leakage and count by reason
leakage_by_reason = (
    df.groupby('denial_reason')
    .agg(
        total_leakage=('billed_amount', 'sum'),
        claim_count=('billed_amount', 'count')
    )
    .sort_values(by='total_leakage', ascending=False)
    .reset_index()
)

#percentage of total leakage column
total_lost = leakage_by_reason['total_leakage'].sum()
leakage_by_reason['%_of_total_leakage'] = (leakage_by_reason['total_leakage'] / total_lost) * 100

# Display the results
display(leakage_by_reason)
```

CORRELATION ANALYSIS

 Top 5 provider-diagnosis combinations with the highest denial rates.


```python
#Group by both Provider and Diagnosis
#Calculate the mean (Denial Rate) and count (Volume)
cohort_analysis = (
    df.groupby(['provider_name', 'icd10_description'])['is_denied']
    .agg(['mean', 'count'])
    .reset_index()
    )

top_5_cohorts = (
    cohort_analysis[cohort_analysis['count'] >= 5]
    .sort_values(by='mean', ascending=False)
    .head(5)
)

print("--- Top 5 Provider-Diagnosis Denial Cohorts ---")
display(top_5_cohorts)
```

DENIAL RISK FLAG

```python
# Create Risk Factors

# High-risk payer
df['risk_payer'] = df['payer_type'].isin(['Government', 'Insurance', 'Self-Pay']).astype(int)
```

```python
# Late processing
threshold_days = df['claim_processing_days'].median()

df['risk_processing'] = (df['claim_processing_days'] > threshold_days).astype(int)
```

```python
# High billed amount
threshold_amount = df['billed_amount'].median()

df['risk_amount'] = (df['billed_amount'] > threshold_amount).astype(int)
```

```python
# Create Risk Score
df['risk_score'] = (
    df['risk_payer'] +
    df['risk_processing'] +
    df['risk_amount']
)
print(df['risk_score'])
```
